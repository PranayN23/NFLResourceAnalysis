# Team Success Prediction

This notebook focuses on predicting Team Wins or EPA (Expected Points Added) based on the aggregated performance metrics of its players.

## Logic
1. **Input**: Roster aggregation (e.g., Average QB PFF, Average OL Grade, WR Separation Avg).
2. **Target**: Team Wins / Playoff Status.
3. **Model**: Dense Neural Network or Simple Regression (given input size).

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Load Team Data (Mocking structure based on project description)
# You would likely need to merge 'Combined_QB.csv' with other position files and 'nfl_win.csv'

# For this template, we assume a 'team_aggregated.csv' exists or we construct it.
# Let's mock the data loading for the structure:

print("Loading aggregated team data...")
# df_team = pd.read_csv('../team_aggregated.csv')

# Example Features:
# QB_Grade, OL_PassBlock, WR_Grade, Defense_Coverage, PassRush_WinRate
features = ['QB_Grade', 'OL_PassBlock', 'WR_Grade', 'Defense_Coverage', 'PassRush_WinRate']
target = 'Wins'

print("Defining Model architecture...")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(len(features),)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)  # Linear output for Wins
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

print("Model compiled. Ready for training once data pipeline is linked.")
```

```python
def train_team_model(X, y):
    # Standard Training Loop
    history = model.fit(X, y, epochs=100, validation_split=0.2, verbose=0)
    return model, history

# Save logic
# model.save('team_prediction_model.keras')
```
