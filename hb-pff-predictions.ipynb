```python
import pandas as pd
import numpy as np
from tensorflow.keras.layers import Input, Dense, SimpleRNN, Dropout, LayerNormalization, MultiHeadAttention
from tensorflow.keras.models import Model
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('/kaggle/input/hbdata/Combined_RB.csv')

# Update the target metric
metric = 'Current_PFF_x'
print(df.columns)

# Sort data by team and year for consistency
df = df.sort_values(by=['Team', 'Year'])
print(df.head())

# Prepare sequences and targets
sequences = []
targets = []

# Group data by team
team_data = df.groupby('Team')

for team, group in team_data:
    if len(group) >= 4:  # Ensure enough data for each team
        print(f"Processing team: {team}, data length: {len(group)}")
        
        # Generate sequences for each team
        for i in range(len(group) - 3):
            # Select input features relevant for PFF prediction
            sequence = group.iloc[i:i+3][[
                'weighted_avg_attempts', 
                'weighted_avg_avoided_tackles', 
                'weighted_avg_breakaway_attempts', 
                'weighted_avg_breakaway_percent', 
                'weighted_avg_breakaway_yards', 
                'weighted_avg_designed_yards', 
                'weighted_avg_drops', 
                'weighted_avg_elu_rush_mtf'
            ]]

            # Update target to predict average PFF metric over the next 3 years
            target = np.mean(group.iloc[i+1:i+4][metric])
            
            sequences.append(sequence.values)
            targets.append(target)

# Convert lists to numpy arrays
X = np.array(sequences)
y = np.array(targets)

# Reshape X to fit RNN input shape
X = X.reshape(X.shape[0], 3, -1)

# Normalize features
scaler = StandardScaler()
X = scaler.fit_transform(X.reshape(-1, X.shape[2])).reshape(X.shape)

print(f"Total sequences generated: {len(sequences)}")
print(f"X shape: {X.shape}, y shape: {y.shape}")

# Define the model
inputs = Input(shape=(3, X.shape[2]))
dense_input = Dense(16, activation='relu')(inputs)

# Multi-head attention layer
attention_output = MultiHeadAttention(num_heads=4, key_dim=X.shape[2])(dense_input, dense_input)
attention_output = LayerNormalization()(attention_output)

# RNN layer to capture temporal patterns
rnn_output = SimpleRNN(units=64, activation='relu', return_sequences=False)(attention_output)
rnn_output = Dropout(0.2)(rnn_output)

# Output layer for predicting PFF
output = Dense(1)(rnn_output)

# Compile the model
model = Model(inputs=inputs, outputs=output)
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model with varying epochs
epoch_list = [98,100,120,150,200]
batch_size = 20

for e in epoch_list:
    print(f"Training for {e} epochs")
    history = model.fit(X, y, epochs=e, batch_size=batch_size, validation_split=0.2)
    
    # Plot training and validation loss
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'Training vs Validation Loss (Epochs: {e})')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    
    # Predictions
    y_pred = model.predict(X)
    
    # R² Score
    r2 = r2_score(y, y_pred)
    print(f"R² Score: {r2:.4f}")
    
    # Plot actual vs predicted PFF
    plt.figure(figsize=(10, 5))
    plt.plot(y, label='Actual', color='blue')
    plt.plot(y_pred, label='Predicted', color='red', linestyle='dashed')
    plt.title(f'Actual vs Predicted Average PFF (Epochs: {e})')
    plt.xlabel('Samples')
    plt.ylabel('Average PFF')
    plt.legend()
    plt.show()

```
